```python
import sys
sys.path.insert(0,'tagger')

import extras, tagger, build_dict
from tagger import Tagger,Stemmer,Reader,Rater
import pandas as pd
import nltk
import cPickle
```

## Image tagging using tagger

```python
df = pd.read_csv('data/Flickr8k_text/Flickr8k.token.txt',sep='\t',names=['image','caption'])
image = [d.split('#')[0] for d in df.image.values]
df.image = image
df
```

```python
extras.build_dict_from_nltk('data/nltk_weight.okl', nltk.corpus.brown,nltk.corpus.stopwords.words('english'), measure='ICF')
```

```python
weights = cPickle.load(open('data/nltk_weight.okl', 'rb')) # or your own dictionary
myreader = tagger.Reader() # or your own reader class
mystemmer = tagger.Stemmer() # or your own stemmer class
myrater = tagger.Rater(weights) # or your own... (you got the idea)
mytagger = Tagger(myreader, mystemmer, myrater)
```

```python
%time tags = [mytagger(d,3) for d in df.caption.values]
```

```python
docs = []
for i,att in enumerate(df.values):
    im = att[0]
    caption = att[1]
    doc = {}
    doc['id'] = 'data/Flicker8k_Dataset/'+ im
    doc['caption'] = caption
    doc['tags'] = tags[i]
    docs.append(doc)
print 'number of documents: ', len(docs)
docs[:5]
```
